```python
emergency_contacts = [
    ("Alice Smith", "555-123-4567"),
    ("Bob Johnson", "bob.johnson@example.com"),
    ("Charlie Brown", "555-987-6543")
]

print("Emergency Contacts:", emergency_contacts)

import array
num_contacts = len(emergency_contacts)
alert_statuses = array.array('i', [0] * num_contacts)

print("Alert Statuses Array:", alert_statuses)

import time

print("SOS Alert System Initialized. Press 'S' to send SOS or 'Q' to quit.")

while True:
    user_input = input("Enter command (S/Q): ").strip().upper()

    if user_input == 'Q':
        print("Exiting SOS Alert System.")
        break

    elif user_input == 'S':
        print("\nSOS Alert Triggered! Simulating sending alerts...")
        for i, (name, contact_info) in enumerate(emergency_contacts):
            print(f"  Simulating alert for {name} ({contact_info})...")
            alert_statuses[i] = 1  # Mark as sent
            time.sleep(0.5)  # Simulate network delay
        print("Alert simulation complete.\n")

    else:
        print("Invalid command. Please press 'S' to send SOS or 'Q' to quit.")
```
