```python
import matplotlib.pyplot as plt
import pyttsx3
import speech_recognition as sr
import webbrowser
import os
import pywhatkit
import wikipedia
import pyautogui
import time


speaker = pyttsx3.init()
mic = sr.Recognizer()

speaker.say("Jarvis welcome you ")
speaker.runAndWait()

with sr.Microphone() as source:
    print("Listening...")
    voice = mic.listen(source)
    command = ""
    try:
        text = mic.recognize_google(voice)
        print(text)
        command = text.lower()
        if "open youtube" in command:
            webbrowser.open("https://www.youtube.com")
        elif "open google" in command:
            webbrowser.open("https://www.google.com")
        elif "open whatsapp" in command:
            os.system("open /Applications/WhatsApp.app")
        elif "open instagram" in command:
            webbrowser.open("https://www.instagram.com")
        elif "open facebook" in command:
            webbrowser.open("https://www.facebook.com")
        elif "open twitter" in command:
            webbrowser.open("https://www.twitter.com")
        elif "open linkedin" in command:
            webbrowser.open("https://www.linkedin.com")
        elif "open github" in command:
            webbrowser.open("https://www.github.com")
        elif "open stackoverflow" in command:
            webbrowser.open("https://www.stackoverflow.com")
        elif "send message" in command:
            try:

                # Define contacts dictionary - USER MUST UPDATE THIS
                contacts = {

                    "friend": "+917878275650"
                }

                speaker.say("What message should I send?")
                speaker.runAndWait()

                with sr.Microphone() as source:
                    print("Listening for message...")
                    voice_msg = mic.listen(source)
                    message = mic.recognize_google(voice_msg)
                    print(f"Message: {message}")

                # For now, let's just send to a hardcoded number or ask for a name if you want
                # To keep it simple based on your previous request, I'll ask for a name
                speaker.say("Who should I send it to?")
                speaker.runAndWait()

                with sr.Microphone() as source:
                    print("Listening for recipient...")
                    voice_recipient = mic.listen(source)
                    recipient_name = mic.recognize_google(voice_recipient).lower()
                    print(f"Recipient: {recipient_name}")

                if recipient_name in contacts:
                    phone_number = contacts[recipient_name]

                    # Construct the URL for the desktop app
                    # This uses the custom URL scheme for WhatsApp
                    url = f"whatsapp://send?phone={phone_number}&text={message}"

                    # Open the URL using the system's 'open' command
                    os.system(f"open '{url}'")

                    # Wait for the app to open and load
                    time.sleep(3)

                    # Press Enter to send
                    pyautogui.press('enter')

                    speaker.say(f"Message sent to {recipient_name}")
                    speaker.runAndWait()
                else:
                    speaker.say("Contact not found.")
                    speaker.runAndWait()

            except Exception as e:
                print(f"Error sending message: {e}")
                speaker.say("Sorry, I encountered an error while searching.")

        elif "take screenshot" in command or "screenshot" in command:
            try:
                timestamp = time.strftime("%Y%m%d-%H%M%S")
                screenshot_path = f"screenshot_{timestamp}.png"
                # Save to Desktop for easier visibility
                desktop_path = os.path.expanduser("~/Desktop")
                full_path = os.path.join(desktop_path, screenshot_path)

                print("Taking screenshot...")
                screenshot = pyautogui.screenshot()
                print(f"Saving screenshot to {full_path}...")
                screenshot.save(full_path)
                print("Screenshot saved successfully.")

                speaker.say("Screenshot taken and saved to Desktop.")
                speaker.runAndWait()
            except Exception as e:
                print(f"Error taking screenshot: {e}")
                import traceback
                traceback.print_exc()
                speaker.say("Sorry, I couldn't take the screenshot.")
                speaker.runAndWait()
    except Exception as e:
        print("Error: " + str(e))
```
